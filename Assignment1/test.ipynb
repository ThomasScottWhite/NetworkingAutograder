```python
import sys
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def import_graph(text):
    with open('./tests/test1/input/1.txt', 'r') as file:
        data = file.read()
    edges = eval(f"[{data}]")
    G = nx.Graph()
    G.add_edges_from(edges)
```

```python
data = sys.stdin.read()
```

```python
plt.figure(figsize=(8, 6))
nx.draw(
    G, 
    with_labels=True, 
    node_color='skyblue', 
    node_size=800, 
    font_size=14, 
    font_weight='bold', 
    edge_color='gray',
    arrows=True,
    arrowsize=20, 
    arrowstyle='-|>'
)
plt.title("Graph Visualization with NetworkX", fontsize=16)
plt.show()
```
